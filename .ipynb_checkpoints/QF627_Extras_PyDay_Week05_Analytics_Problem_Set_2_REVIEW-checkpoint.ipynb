> Another way to analyse how default rates vary across continous variables is to bin the variables into discrete categories.

> Let's create a categorical variable for the loan amount variable into small, medium, high, very high.

* loan_amount < 5000 `low`
* loan_amount >=5000 and loan_amount < 15000 `medium`
* loan_amount >= 15000 and loan_amount < 25000 `high`
* greater than 25000 `very high`

````python
loan_status_cleaning["loan_amnt_categories"] =\
(
    pd
    .cut(x = bank["loan_amnt"],
         bins = [0, 5000, 15000, 25000, np.inf], # intervals
         labels = ["low", "medium", "high", "very high"], # labels should be bins - 1
         right = False
        )
)

loan_status_cleaning.head(3).T
````

````python
loan_status_cleaning[["loan_amnt_categories", "loan_amnt"]]
````

````python
subset_bank_with_default_rate_3 =\
(
    loan_status_cleaning
    .groupby("loan_amnt_categories")
    ["loan_status"]
    .agg(defaults = "sum",
         total = "count" # COUNT
        )
)
````

````python
subset_bank_with_default_rate_3["default_rate"] =\
(
    subset_bank_with_default_rate_3["defaults"]
    /
    subset_bank_with_default_rate_3["total"]
    * 100
)

subset_bank_with_default_rate_3 =\
(
    subset_bank_with_default_rate_3
    .reset_index()
)
````

````python
subset_bank_with_default_rate_3
````

````python
(
    ggplot(subset_bank_with_default_rate_3,
           aes(x = "loan_amnt_categories",
               y = "default_rate",
               fill = "loan_amnt_categories")
          )
    + geom_bar(stat = "identity")
    + theme(legend_position = "none") 
    + labs(x = "loan_amnt_categories",
           y = "Loan Default Rate",
           title = "Voila!")
)
````

#### debt-to-income (`dti`) ratio

````python
%whos
````

> Please create a categorical variable of `dti` using your own function

> Use the following definitions for each categories.

* <= 10 `low`
* dti > 10 and dti <=20 `medium`
* greater than 20 `high`

````python
# for your conveneint use

def cat_var(x,
            bins,
            labels):
    for bound, label in zip(bins, labels):
        if x <= bound:
            break
    return label 

# see below for use cases at `installment`
````

````python
loan_status_cleaning["dti_categories"] =\
(
    loan_status_cleaning["dti"]
    .apply(lambda x: 
           cat_var(x,
                   [0, 10, 20, np.inf],
                   ["low", "medium", "high"]
                  )
          )
)

loan_status_cleaning.head(3).T
````

````python
subset_bank_with_default_rate_4 =\
(
    loan_status_cleaning
    .groupby("dti_categories")
    ["loan_status"]
    .agg(defaults = "sum",
         total = "count" # COUNT
        )
)
````

````python
subset_bank_with_default_rate_4["default_rate"] =\
(
    subset_bank_with_default_rate_4["defaults"]
    /
    subset_bank_with_default_rate_4["total"]
    * 100
)

subset_bank_with_default_rate_4 =\
(
    subset_bank_with_default_rate_4
    .reset_index()
)
````

````python
subset_bank_with_default_rate_4
````
